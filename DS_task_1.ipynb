```python
import zipfile
import os
import pandas as pd
import matplotlib.pyplot as plt

# Define the path to the uploaded zip file
zip_file_path = '/content/API_SP.POP.TOTL_DS2_en_csv_v2_3401680.zip'
extraction_dir = '/content/extracted_files/'

# Extract zip file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extraction_dir)

# Load the main data file
data_file_path = os.path.join(extraction_dir, 'API_SP.POP.TOTL_DS2_en_csv_v2_3401680.csv')
data = pd.read_csv(data_file_path, skiprows=4)

# Process the 2023 data
population_2023 = data[['Country Name', '2023']].dropna()
population_2023 = population_2023.sort_values(by='2023', ascending=False).head(10)

# Create the bar chart
plt.figure(figsize=(10, 6))
plt.barh(population_2023['Country Name'], population_2023['2023'], color='skyblue')
plt.xlabel('Population in 2023')
plt.ylabel('Country')
plt.title('Top 10 Most Populous Countries in 2023')
plt.gca().invert_yaxis()  # Invert y-axis to have the most populous country at the top
plt.show()
```
